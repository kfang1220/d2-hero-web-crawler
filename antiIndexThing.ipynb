```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
heroDict = pd.read_csv('G:\Dropbox\Dropbox\Python\Output\HeroListWithURL.csv').to_dict('records')

#define converter for percentage string to float
def p2f(x):
    return float(x.strip('%'))

bigdf = []
```

```python
#Set site to be in English
s = requests.session()
setLanguage = s.get('http://dotamax.com/set_language/?language=en') 

#need to define hero
for item in heroDict:
    
    print('scraping {}'.format(item['hero']))
    cont = s.get('http://dotamax.com/hero/detail/match_up_anti/' + item['url'] + '/?skill=vh&ladder=y').text
    soup = BeautifulSoup(cont, 'lxml')

    #Find and target second table which is the table we want
    table = soup.find_all('table')[1]

    #Define dataTable
    dataTable = []

    #Obtain data from rows
    for row in table.find_all('tr'):
        heroData = []
        for a in row.find_all('a'): #obtain image src
            heroData.append(a.img['src'])
        for cell in row.find_all('td'): #obtain data
            heroData.append(cell.text)
        dataTable.append(heroData)

    df = pd.DataFrame(dataTable, columns=['img src', 'vs Hero', 'Anti Index', 'Win Rate', 'Matches Played']) #Build Data Table
    df['Anti Index'] = df['Anti Index'].apply(p2f) #Change percentage to float
    df['Win Rate'] = df['Win Rate'].apply(p2f) #Change percentage to float
    df['hero'] = item['hero']
    
    bigdf.append(df)
```

```python
pd.concat(bigdf).to_csv('PythonProjects/bigdf.csv', index=False)
bigdf = pd.concat(bigdf)
bigdf
```

```python
#Sorts the list alphabetically
sortedList = pd.DataFrame(heroDict) 
sortedList = sortedList.sort_values(by=['hero'])
sortedList = sortedList.values.tolist()

print(sortedList)
```

```python
# write-html.py
newdf = pd.read_csv('bigdf.csv')
newdf['img src'] = newdf['img src'].apply(lambda x: "{}{}{}".format('<img src="', x, '" class="heroImage">'))
newdf.rename(columns={'img src': '', 'vs Hero':''}, inplace=True)

pd.set_option('display.max_colwidth', -1)

newdf
```

```python
f = open('antiIndexBooklet.html','w')

htmlPage = """    
<!DOCTYPE html>
<html>
  <head>
    <meta charset="utf-8">
    <link href="style.css" rel="stylesheet">
    <title>Anti Index Booklet</title>
    <meta name="description" content="Anti Index Booklet">
    <style>
    
	@page {
		size: 5.5in 8.5in;
	}
	
	@media print
	{
	article {page-break-before:always;}
	}
    

* {
    padding: 0;
    margin: 0;
	font-family: 'Open Sans', sans-serif;
}

article {
    width: 21cm;
    height: 27cm;
    page-break-before:always;
    padding: 2cm;
}

.heroLogo {
	height: 60px;
}

table {
    display: table;
    border-collapse: separate;
    border-spacing: 0;
    border: none;
}

.heroImage {
    height:25px;
}

th {
    text-align: left;
}

tr {
    height: 28px;
}

td {
    font-size: 13px;
}

.DataTable {
  float: left;
  width: 40%;
}

/* Clear floats after the columns */
article:after {
  content: "";
  display: table;
  clear: both;
}
</style>
  </head>

  <body>
  """

#Start loop that goes through every hero alphbetically 
for hero in sortedList:
    htmlPage = htmlPage + "<article>\n"
    htmlPage = htmlPage + """<img src="http://cdn.dota2.com/apps/dota2/images/heroes/""" + hero[1] + """_hphover.png" class="heroLogo"><h1>""" + hero[0] + """</h1>"""
    #htmlPage = htmlPage + """<h2>Top 30</h2>"""
    htmlPage = htmlPage + newdf[newdf['hero'] == hero[0]].sort_values('Anti Index', ascending=False).drop(newdf.columns[[3,4,5]], axis=1)[:30].to_html(classes="DataTable", border=0, index=False, escape=False) #outputs the top 30 anti index 
    #htmlPage = htmlPage + """<h2>Bottom 30</h2>"""
    htmlPage = htmlPage + newdf[newdf['hero'] == hero[0]].sort_values('Anti Index', ascending=False).drop(newdf.columns[[3,4,5]], axis=1)[-30:].to_html(classes="DataTable", border=0, index=False, escape=False) #outputs the bottom 30 anti index
    htmlPage = htmlPage + "</article>"
    
    #ends the page and goes to the next hero

htmlPage = htmlPage + """"</body></html>"""

f.write(htmlPage)
f.close()
```
